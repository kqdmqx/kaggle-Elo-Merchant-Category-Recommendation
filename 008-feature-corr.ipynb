```python
from data_io import load_train_newk_features
```

```python
part = load_train_newk_features(20000)
```

```python
part.head().T
```

```python
corr_mat = part.corr()
```

```python
corr_col = corr_mat.loc["newk_new_card_id_count"]
corr_col[corr_col > 0.998]
```

```python
corr_col = corr_mat.loc["newk_new_merchant_id_nunique"]
corr_col[corr_col > 0.998]
```

```python
dup_cols = (corr_mat > 0.998).sum() 
dup_cols = dup_cols[dup_cols > 1]
```

```python
to_remove = []
for col in dup_cols.index:
    if col in to_remove:
        continue
    corr_col = corr_mat.loc[col]
    to_remove += [col_ for col_ in corr_col[corr_col > 0.998].index if col_ != col]

print(len(to_remove))
to_remove
```
